## Loading the JSON file

Load the data from the file `disease_data.json`.

```python
import pandas as pd
import json
```

```python
f = open('disease_data.json')
data = json.load(f)
```

## Explore the first and second levels of the schema hierarchy

```python
data.keys()
```

```python
print('meta: ', type(data['meta']))
print('data: ', type(data['data']))
```

```python
# investigate the 'meta' key of the overall data dict which apparently holds a dict w/ a single key
print(data['meta'].keys())
```

```python
print('view: ', type(data['meta']['view']))
```

```python
data['meta']['view'].keys() # appears to contain info about the json file itself
```

```python
for key in data['meta']['view'].keys():
    print(key, type([data['meta']['view'][key]])) # all the keys in 'view' hold lists
```

```python
# investigate the 'data' key of the overall data dict which holds a list
# check out what the first entry of the list looks like:
print(type(data['data'][0])) # each entry is apparently a list
data['data'][0]
```

```python
data['data'][1]
# so each entry in the list is another list that can't be broken down any further
# looks like this is the level that holds the data we'll be interested in, but it's unclear what any of it means
##  until we match it up with column names from within the 'meta' key
```

```python
len(data['data'][1]) # looks like we need 42 col names
```

## Convert to a DataFrame

Create a DataFrame from the JSON file. Be sure to retrive the column names for the dataframe. (Search within the 'meta' key of the master dictionary.) The DataFrame should include all 42 columns.

```python
df = pd.DataFrame(data['data'])
df.head()
```

```python
# make a list of col names to add to above df
col_names = []
for col in data['meta']['view']['columns']:
    col_name = col['name']
    col_names.append(col_name)
col_names
```

```python
df.columns = col_names
df.head()
```

```python
# could have accomplished above with a list comprehension
df = pd.DataFrame(data['data'])
print(df.shape)
df.columns = [col['name'] for col in data['meta']['view']['columns']]
print(df.columns)
df.head()
```
